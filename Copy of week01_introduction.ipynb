Broadcasting <a class="anchor" id="broadcasting"></a>
-----------------------------------------------------

Note that the dot product and the “*" operator for arrays is different.
That means that using the "*” operator does
[broadcasted](https://docs.scipy.org/doc/numpy/user/theory.broadcasting.html#array-broadcasting-in-numpy)
(the smaller array is repeated to fit the size of the larger array)
element-wise multiplication instead of the dot operator:

    ones = np.ones((3, 1))
    R = np.random.normal(size=(3,3))
    print("ones:\n", ones)
    print("R:\n", R)
    print("R * ones:\n", R * ones)
    print("dot(R, ones):\n", np.dot(R, ones))

```python
ones = np.ones((3, 1))                  # print 3 rows and 1 column of 'ones'
R = np.random.normal(size=(3,3))        # random normal distribution matrix (3x3)
print("ones:\n", ones)
print("R:\n", R)
print("R * ones:\n", R * ones)          # elementwise multiplikation; 
print("dot(R, ones):\n", np.dot(R, ones))     # row*column multiplik; 

# your code here
```
